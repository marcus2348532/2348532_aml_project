```python
import pandas as pd
storm=pd.read_csv("/content/climate_disaster_storm - Storm.csv")
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder();
storm["Indicator"]=le.fit_transform(storm["Indicator"])
country=storm["Country"]
x=0
for i in storm.columns:
  if x<10:
    storm.drop(i,axis=1,inplace=True)
    x+=1
storm_india=pd.DataFrame()
storm_india=storm.T
storm_india.columns=country
import numpy as np
storm_india.replace(np.nan, 0, inplace=True)
x=range(1980,2023)
storm_india["Years"]=x
x_storm=storm_india["India"]
y_storm=storm_india["Years"]
storm_india.reset_index(inplace=True)
storm_india
```

```python
temperature=pd.read_csv("/content/Extreme temperature.xlsx - Extreme temperature.csv")
temperature["Indicator"]=le.fit_transform(temperature["Indicator"])
country=temperature["Country"]
x=0
for i in temperature.columns:
  if x<11:
    temperature.drop(i,axis=1,inplace=True)
    x+=1
df_temp=pd.DataFrame()
df_temp=temperature.T
df_temp.columns=country
df_temp.replace(np.nan, 0, inplace=True)
x=range(1980,2023)
df_temp["Years"]=x
x_temp=df_temp["India"]
y_temp=df_temp["Years"]
df_temp.reset_index(inplace=True)
df_temp
```

```python
import matplotlib.pyplot as plt
plt.plot(y_storm,x_storm,marker="o",c='b')
plt.plot(y_temp,x_temp,marker="o",c='r')
plt.legend({"storm","extreme temperature"})
plt.show()
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
x=main[['year']]
y=main['globalWarmin_lowess']
lr=LinearRegression()
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=11)
lr.fit(xtrain,ytrain)
LRypred=lr.predict(xtest)
maeLR = mean_absolute_error(ytest,LRypred)
mseLR = mean_squared_error(ytest,LRypred)
r2LR = r2_score(ytest,LRypred)
LR=[maeLR,mseLR,r2LR]
plt.scatter(xtrain,ytrain,c='r')
plt.scatter(xtest,LRypred,c='b')
plt.legend({'actual'})
plt.xlabel('YEAR')
plt.ylabel('globalWarmin_lowess')
print("LINEAR REGRESSION")
print("Mean Absolute Error:", maeLR)
print("Mean Squared Error:", mseLR)
print("R2 Score:", r2LR)
```

```python
from sklearn.preprocessing import PolynomialFeatures
p2 = PolynomialFeatures(degree = 2, include_bias = True)
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=0.20,random_state=2)
xtrainp = p2.fit_transform(xtrain)
xtestp =p2.transform(xtest)
model = LinearRegression()
model.fit(xtrainp,ytrain)
ypred = model.predict(xtestp)
plt.scatter(xtrain,ytrain,c='r')
plt.scatter(xtest,ytest,c='b')
plt.legend({'actual'})
plt.xlabel('YEAR')
plt.ylabel('globalWarmin_noSMoothe')
r2PR = r2_score(ytest, ypred)
msePR = mean_squared_error(ytest, ypred)
maePR = mean_absolute_error(ytest, ypred)
PR=[maePR,msePR,r2PR]
print("CO-EFFICIENT:",model.coef_)
print("INTERCEPT:",model.intercept_)
print(f'R-squared (R2): {r2PR:.4f}')
print(f'Mean Squared Error (MSE): {msePR:.4f}')
print(f'Mean Absolute Error (MAE): {maePR:.4f}')
```
